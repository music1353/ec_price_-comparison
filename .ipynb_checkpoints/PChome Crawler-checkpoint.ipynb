```python
import requests
import json
import time
```

```python
url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=ssd&page=1&sort=sale/dc'
r = requests.get(url)

prod_list = []
if r.status_code == 200:
    data = json.loads(r.text)
    
    # 取出product data
    for item in data['prods']:
        PROD_BASE_URL = 'https://24h.pchome.com.tw/prod/'
        IMG_BASE_URL = 'https://b.ecimg.tw'
        
        url = PROD_BASE_URL + item['Id']
        name = item['name']
        img_url = IMG_BASE_URL + item['picS']
        price = item['price']
        
        obj = {
            'url': url,
            'name': name,
            'img_url': img_url,
            'price': price
        }
        prod_list.append(obj)
        
print(len(prod_list))
prod_list
```

```python
prod_list = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

for page_number in range(1, 6):
    print(page_number)
    url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=ssd&page='+ str(page_number) +'&sort=sale/dc'
    r = requests.get(url, headers=headers)
    
    if r.status_code == 200:
        data = json.loads(r.text)

        # 取出product data
        for item in data['prods']:
            PROD_BASE_URL = 'https://24h.pchome.com.tw/prod/'
            IMG_BASE_URL = 'https://b.ecimg.tw'

            url = PROD_BASE_URL + item['Id']
            name = item['name']
            img_url = IMG_BASE_URL + item['picS']
            price = item['price']

            obj = {
                'url': url,
                'name': name,
                'img_url': img_url,
                'price': price
            }
            prod_list.append(obj)
            
        time.sleep(2)
        
        if page_number == 6:
            print(r.status_code)
            print('已搜索到' + str(page_number) + '頁')
            print('已搜索到' + str(len(prod_list)) + '樣商品')
    else:
        print(r.status_code)
        print('已搜索到' + str(page_number) + '頁')
        print('已搜索到' + str(len(prod_list)) + '樣商品')
        break
```
