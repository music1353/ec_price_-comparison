```python
SUNSUN_URL = 'https://www.u-mall.com.tw/Search/Get'
SUNSUN_PROD_LINK_URL = 'https://www.u-mall.com.tw/'

POST_HEADER = {
    'keyword': 'iphone',
    'model[cateName]': '全站',
    'model[pageSize]': '100'
}

prod_list = []
if resp.status_code == 200:
    resp = requests.post(SUNSUN_URL, data=POST_HEADER)
    ori_data = json.loads(resp.text)
    data = ori_data['searchResult']['products']
    
    for item in data:
        prod_id = str(item['id'])
        prod_name = item['title']
        prod_price = item['finalPrice']
        prod_url = SUNSUN_PROD_LINK_URL + item['pageLink']
        prod_pic = item['imageUrl']

        obj = {
            'id': prod_id,
            'name': prod_name,
            'prod_price': prod_price,
            'url': prod_url,
            'prod_pic': prod_pic
        }
        prod_list.append(obj)
else:
    print('失敗')
    
print(prod_list)
print(len(prod_list))
```
